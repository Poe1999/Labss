```python
import matplotlib.pylab as plt
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, color = 'darkred', linestyle = '--', label = 'sin(x)')
plt.plot(x,z, color = 'purple', linestyle = ':', label = 'cos(x)')
plt.title('  ', color = 'darkred')
plt.xlabel('Ox')
plt.ylabel('Oy')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
x = np.random.normal(loc = 0, scale=1, size=1000)
plt.hist(x, bins = 20, color = 'darkred')
plt.title(' ', color = "darkred")
plt.xlabel('')
plt.ylabel('')

plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
x = np.random.randint(1,10,(5,5))
avg = np.mean(x)
max_el = np.max(x)
min_el = np.min(x)
suum = np.sum(x,axis=0)
print(x)
print(avg)
print(max_el)
print(min_el)
print(suum)
```

```python
import matplotlib.pylab as plt
import numpy as np
x = np.random.randint(1,10,(5,5))
plt.imshow(x, cmap='autumn')
plt.colorbar()
for i in range(5):
    for j in range(5):
        plt.text(i,j,f'{x[i,j]}')
plt.title(' ', color = 'darkred')
plt.show()
```

```python
import matplotlib.pylab as plt
import numpy as np
fig, ax = plt.subplots(1,3)

x = np.linspace(-10,10,100)
ax[0].plot(x, x**2, color = 'darkred', linestyle = '--')
ax[0].set_title('y=x^2')

x1= np.random.rand(50)
y1= np.random.rand(50)
ax[1].scatter(x1,y1, color = 'darkred')
ax[1].set_title(' ')


x3 = ['P','L','M','F','A']
y3 = [176,169,168,162,156]
colors = ['#8B0000','#FFD700','#FF1493','#8B008B','#87CEEB']
ax[2].bar(x3,y3,color = colors )
ax[2].set_title(' ')

plt.show()
```
